```python
df_US=df[df["Country"]=='US']
df_US
```

```python
df_US["Last Update"]=df_US["Last Update"].str[:10]
df_US["Last Update"]=pd.to_datetime(df_US["Last Update"])
```

```python
df_US = df_US.groupby("Last Update").sum().reset_index()
df_US
```

```python
df_US["Country Name"]="United States"
df_US.drop("Sno",inplace=True,axis=1)
```

```python
df_US.head()
```

### Let's include some external dataset

1. Countrywise Morality rate 
2. Countrywise Population density 
3. Countrywise Population
4. Transmission Rate we are using the same as we used for china.

```python
health_rate=pd.read_csv("healthrating.csv")[["Country Name","2016"]]
population=pd.read_csv("population.csv")[["Country Name","2018"]]
population_density=pd.read_csv("country_density.csv")[["Country Name","2018"]]
```

```python
df_US = pd.merge(df_US, health_rate, how='left', on='Country Name')
df_US = pd.merge(df_US, population, how='left', on='Country Name')
df_US.rename(columns = {"2016":"Morality rate","2018":"population"},inplace=True)
df_US = pd.merge(df_US, population_density, how='left', on='Country Name')
df_US.rename(columns = {"2018":"population_density"},inplace=True)
df_US["Tansmission_rate"]=2.68/df_US["Morality rate"]
df_US.drop("Morality rate",inplace=True,axis=1)
```

```python
df_US["Last Update"]=df_US["Last Update"]+ pd.DateOffset(1)
```

```python
df_US.rename(columns = {"Confirmed":"y","Last Update":"ds"},inplace=True)
```

```python
def cal_US(df):
    test=df.tail(2)
    df=df.head(8)
    from fbprophet import Prophet
    m = Prophet()
    m.fit(df)
    test.drop("y",inplace=True,axis=1)
    forecast = m.predict(test)
    #forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
    print(forecast["yhat"])
```

```python
cal_US(df_US)
```
